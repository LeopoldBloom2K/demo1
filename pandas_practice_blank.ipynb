## 🔍 Chapter 2 — Indexing, Selecting & Assigning
**목표**: iloc / loc / 조건부 선택 자유자재로 다루기

```python
# ─────────────────────────────────────────
# Q1. iloc으로 df의 마지막 3행, 처음 4개 컬럼을 선택하세요.
# ─────────────────────────────────────────
# 힌트: df.iloc[시작:끝, 시작:끝]  ← 음수 인덱스(-3) 사용 가능

result = df.iloc[-3, :4]

print('[Q1 결과]')
print(result)
```

```python
# ─────────────────────────────────────────
# Q2. loc으로 order_id 1005~1008번의
#     customer_name, product, price 컬럼만 선택하세요.
# ─────────────────────────────────────────
# 힌트: df.loc[시작id:끝id, ['컬럼1', '컬럼2', '컬럼3']]
#       ★ loc은 끝 레이블 포함!

result = df.loc[1005:1008, ['customer_name', 'product', 'price']]

print('[Q2 결과]')
print(result)
```

```python
# ─────────────────────────────────────────
# Q3. 아래 두 조건을 모두 만족하는 행을 필터링하세요.
#     조건 1) category가 '전자기기'
#     조건 2) rating이 4.5 이상
#     출력: product, price, rating 컬럼만
# ─────────────────────────────────────────
# 힌트: df.loc[(조건1) & (조건2), ['컬럼'...]]
#       ★ 각 조건은 반드시 ()로 감쌀 것!

result = df.loc[(df['category'] == '전자기기') & (df['rating'] >= 4.5), ['product', 'price', 'rating']]

print('[Q3 결과]')
print(result)
```

```python
# ─────────────────────────────────────────
# Q4. df에 'total_amount' 컬럼을 추가하세요.
#     계산식: price × quantity
# ─────────────────────────────────────────
# 힌트: df['새컬럼명'] = df['컬럼A'] * df['컬럼B']

df['total_amount'] = df['price'] * df['quantity']

print('[Q4 결과] total_amount 상위 5행')
print(df[['product', 'price', 'quantity', 'total_amount']].head(5))
```

## 📊 Chapter 3 — Summary Functions and Maps
**목표**: 데이터 요약 & map/apply로 전체 변환하기

```python
# ─────────────────────────────────────────
# Q1. price 컬럼의 요약 통계를 출력하세요.
#     추가로 category 컬럼의 값별 개수도 출력하세요.
# ─────────────────────────────────────────
# 힌트: df['컬럼'].describe()  /  df['컬럼'].value_counts()

print('[Q1-1] price 요약 통계')
df['price'].describe()

print('\n[Q1-2] category 값별 개수')
df['category'].value_counts()
```

```python
# ─────────────────────────────────────────
# Q2. map을 사용해 price를 만원 단위로 변환하여
#     'price_만원' 컬럼을 만드세요.
#     (소수점 1자리, 예: 1200000 → 120.0)
# ─────────────────────────────────────────
# 힌트: df['컬럼'].map(lambda x: round(x / 10000, 1))

df['price_만원'] = df['price'].map(lambda x: round(x / 10000, 1))

print('[Q2 결과]')
print(df[['product', 'price', 'price_만원']].head(5))
```

```python
# ─────────────────────────────────────────
# Q3. apply를 사용해 각 행의 total_amount를 기준으로
#     'vip' 컬럼을 만드세요.
#     조건: total_amount >= 200000 → True, 미만 → False
# ─────────────────────────────────────────
# 힌트: df.apply(lambda row: ___, axis=1)
#       row['컬럼명'] 으로 해당 행의 값에 접근

df['vip'] = df.apply(lambda row: row['total_amount'] >= 200000, axis=1)

print('[Q3 결과]')
print(df[['product', 'total_amount', 'vip']].head(8))
```

## 🧹 Chapter 5 — Data Types and Missing Values
**목표**: 타입 변환 & 결측치 탐지·처리 완전 정복

```python
# ─────────────────────────────────────────
# Q1. df 전체 컬럼의 타입을 확인하고,
#     결측치가 있는 컬럼과 그 개수를 출력하세요.
# ─────────────────────────────────────────
# 힌트: df.dtypes
#       df.isnull().sum()  ← 0보다 큰 것만 필터: [df.isnull().sum() > 0]

print('[Q1-1] 컬럼별 데이터 타입')
df.dtypes

print('\n[Q1-2] 결측치 있는 컬럼')
df.isnull().sum() > 0
```

```python
# ─────────────────────────────────────────
# Q2. rating 컬럼의 결측치를 중앙값(median)으로 채우세요.
#     채우기 전과 후의 결측치 개수를 각각 출력하세요.
# ─────────────────────────────────────────
# 힌트: df['컬럼'].isnull().sum()  ← 결측치 개수
#       df['컬럼'] = df['컬럼'].fillna(df['컬럼'].median())

print(f'[Q2] fillna 전 결측치 수: {df["rating"].isnull().sum()}')

df['rating'].isnull().sum()
df['rating'] = df['rating'].fillna(df['rating'].isnull().median())

print(f'[Q2] fillna 후 결측치 수: {df["rating"].isnull().sum()}')  # 0 이어야 정답!
```

```python
# ─────────────────────────────────────────
# Q3-1. is_returned 컬럼을 bool에서 int64로 변환하여
#        'is_returned_int' 컬럼을 만드세요.
# Q3-2. region 컬럼에서 '서울'을 'SEOUL'로 일괄 치환하세요.
# ─────────────────────────────────────────
# 힌트 Q3-1: df['컬럼'].astype('타입문자열')  → 'int64'
# 힌트 Q3-2: df['컬럼'].replace('기존값', '바꿀값')

# Q3-1
df['is_returned_int'] = df['is_returned'].astype('int64')

# Q3-2
df['region'] = df['region'].replace('서울', 'SEOUL')

print('[Q3-1] bool → int 변환')
print(df[['is_returned', 'is_returned_int']].head(6))
print('\n[Q3-2] 서울 → SEOUL 치환')
print(df['region'].unique())
```

## 🔗 Chapter 6 — Combining
**목표**: concat / merge / join으로 데이터 결합하기

```python
# ─────────────────────────────────────────
# Q1. df를 절반으로 나눈 뒤 concat으로 다시 합치세요.
#     합친 결과의 shape을 출력해 원본과 같은지 확인하세요.
# ─────────────────────────────────────────
# 힌트: df.iloc[:10]  /  df.iloc[10:]
#       pd.concat([df1, df2], axis=0)

first_half  = df.iloc[:10]
second_half = df.iloc[10:]
combined    = pd.concat([first_half, second_half], axis=0)

print(f'[Q1] 원본 shape: {df.shape}  |  합친 shape: {combined.shape}')  # 동일해야 정답!
```

```python
# ─────────────────────────────────────────
# Q2. df(주문)와 customers(고객 정보)를
#     'customer_name'을 기준으로 LEFT JOIN 하세요.
#     결과에서 customer_name, product, price, age, membership 컬럼만 출력
# ─────────────────────────────────────────
# 힌트: df.reset_index().merge(other_df, on='공통컬럼', how='___')
#       LEFT JOIN → how='left'

merged = df.reset_index().merge(customers, on='customer_name', how='left')

print('[Q2 결과] 상위 5행')
print(merged[['customer_name', 'product', 'price', 'age', 'membership']].head(5))
```

```python
# ─────────────────────────────────────────
# Q3. merge된 데이터에서 membership 등급별
#     평균 구매금액(total_amount)을 내림차순으로 출력하세요.
#     (소수점 없이 정수로 반올림)
# ─────────────────────────────────────────
# 힌트: merged.groupby('___')['___'].mean()
#              .sort_values(ascending=False).round(0)

result = merged.groupby('membership')['total_amount'].mean().sort_values(ascending=False).round(0)

print('[Q3 결과] 멤버십 등급별 평균 구매금액')
print(result)
```
